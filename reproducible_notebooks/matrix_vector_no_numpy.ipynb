```python
# ------------------------------------------------------------------------------
# Aufgabe: einmal Matrix Multiplikation per Hand OHNE numpy (aber mit python)!
# Ziel: VerstÃ¤ndnis der linearen Algebra-Grundlagen fÃ¼r Deep Learning.
# ------------------------------------------------------------------------------

# --------------------------------------------------
# ðŸ‘‰ 2x2 Matrix
# Ã¤uÃŸere Liste = Zeilen
# innere Liste = SpaltenMatrix
# --------------------------------------------------
matrix = [[1,2],[3,4]]

# --------------------------------------------------
# ðŸ‘‰ Vector mit 2 Elementen
# passend zur Anzahl der Spalten der Matrix!
# --------------------------------------------------
vector = [5,6]

# --------------------------------------------------
# ðŸ‘‰ Ergebnis-Vektor
# LÃ¤nge = Anzahl der Zeilen der Matrix
#     zu erwartendes Ergebnis:
#     -----------------------
#    (1x5)+(2x6) = 17
#    (3x5)+(4x6) = 39
#       --> result = [17,39]
# --------------------------------------------------
result = [0,0] # 2D

# --------------------------------------------------
# ðŸ‘‰ Ã„uÃŸere Schleife â€“ Ã¼ber die Zeilen
# index â†’ Zeilennummer (0, 1)
# value â†’ Matrixwert (z. B. 1, dann 2)
# --------------------------------------------------
print('Value:')
print('------')
for index,row in enumerate(matrix):
	# -----------------------------------------
	# sum = 0   # lieber nicht sum verwenden
	# ðŸ”¹ sum Ã¼berschreibt eine Python-Funktion
	#------------------------------------------
	row_sum = 0
	# --------------------------------------------------
	# ðŸ‘‰ Innere Schleife â€“ Ã¼ber die Spalten
	# col â†’ Spaltenindex (0 oder 1)
	# row â†’ aktuelle Zeile, z. B. [1,2]
	# --------------------------------------------------
	for col,value in enumerate(row):
		print(value)
		# sum += value * vector[col]      # nicht sum - siehe ErklÃ¤rung ðŸ”¹
		row_sum += value * vector[col]    # oder doch zeile?? ;)  -> nein, col ist richtig
	# --------------------------------------------------
	# Ergebnis speichern
	# --------------------------------------------------
	# result[index] = sum
	result[index] = row_sum

print()
print('Result:')
print('------')
print(result)

# Per Hand:
# matrix*vector
print()
print(' âœ… sauber durchgelaufen')
```
